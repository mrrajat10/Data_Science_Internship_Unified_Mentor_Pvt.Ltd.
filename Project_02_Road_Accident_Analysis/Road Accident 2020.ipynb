## **Step-1->Loading, cleaning and EDA The Data:**

```python
# 📦 Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 📥 Load the dataset
df = pd.read_csv(r"C:\Users\Admin\Documents\DataScience\Projects\Datasets\Regulatory Affairs of Road Accident Data 2020 India (1).csv")  

# 🧹 Data Cleaning
df.dropna(subset=['Count'], inplace=True)
df['Count'] = df['Count'].astype(int)

# 📌 Check basic info (optional)
print("Unique Cities:", df['Million Plus Cities'].nunique())
print("Cause Categories:", df['Cause category'].unique())
print("Outcome Types:", df['Outcome of Incident'].unique())

# 📊 1. Total Accidents by Cause Category
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df,
    x='Cause category',
    y='Count',
    estimator=sum,
    errorbar=None,
    palette='coolwarm', 
    legend=False  )       

plt.title("Total Accidents by Cause Category")
plt.xlabel("Cause Category")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

## **📊 Total Accidents by Cause Category:** 
**Distribution:** Most causes (Traffic Control, Junction, Weather, etc.) show accident counts near 175,000, while "Impacting Vehicle/Object" is noticeably lower (~125,000).<br>
**Implication:** Systemic issues like infrastructure and regulation drive most accidents, not isolated collisions. <br>
 **Use case:** Highlights priority areas for road safety interventions and policy focus.

```python
# 📊 2. Top 10 Cities by Total Accidents
top_cities = df.groupby('Million Plus Cities')['Count'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities.index, y=top_cities.values, palette='magma')
plt.title("Top 10 Million-Plus Cities by Total Accidents")
plt.xlabel("City")
plt.ylabel("Total Accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## **🏙️ Top 10 Million-Plus Cities by Total Accidents:** 
**Distribution:** Chennai leads in accident count, followed by Delhi and Jabalpur, while Mumbai ranks lowest among the top 10.<br> 
**Implication:** Urban traffic density does not always correlate directly with accident volume.<br> 
**Use case:** Identifies high-risk cities for targeted traffic safety programs.<br>

```python

# 📊 3. Heatmap: City vs Cause Category
pivot_table = df.pivot_table(index='Million Plus Cities', columns='Cause category', values='Count', aggfunc='sum', fill_value=0)

plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=0.5)
plt.title("Heatmap: City vs Cause Category")
plt.xlabel("Cause Category")
plt.ylabel("Million Plus Cities")
plt.tight_layout()
plt.show()


```

## **🔥 Heatmap: City vs Cause Category:**
 **Distribution:** Jabalpur and Bengaluru show consistently high counts across all causes, while Agra and Coimbatore are lower.<br> 
 **Implication:** Accident causes vary by city, requiring localized strategies.<br>
  **Use case:** Useful for city-specific traffic planning and enforcement.<br>

```python
# 📊 4. Outcome Distribution (All Cities Combined)
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Outcome of Incident', y='Count', estimator=sum, ci=None, palette='pastel')
plt.title("Distribution of Accident Outcomes in Million-Plus Cities (2020)")
plt.xlabel("Outcome")
plt.ylabel("Number of Cases")
plt.tight_layout()
plt.show()
```

## **⚠️ Distribution of Accident Outcomes:**
 **Distribution:** "Total number of Accidents" exceeds 350,000, followed by "Total Injured" (~255,000). Fatalities are ~80,000.<br>
  **Implication:** Minor injuries dominate, but fatalities remain significant.<br> 
  **Use case:** Supports emergency preparedness and healthcare resource allocation.

## **Step-2->Encoding Data for Data Visualization:**

```python

#Cleaning data
# 📦 Import libraries
import pandas as pd

# 📥 Load the dataset
df = pd.read_csv(r"C:\Users\Admin\Documents\DataScience\Projects\Datasets\Regulatory Affairs of Road Accident Data 2020 India (1).csv")

# 🔍 Step 1: View basic info
print("Before Cleaning:")
print(df.info())
print("\nMissing values:\n", df.isnull().sum())

# 🧹 Step 2: Drop rows where 'Count' is missing
df.dropna(subset=['Count'], inplace=True)

# 🧼 Step 3: Fix data types
df['Count'] = df['Count'].astype(int)

# 🧹 Step 4: Remove leading/trailing spaces in string columns
df.columns = df.columns.str.strip()
string_cols = ['Million Plus Cities', 'Cause category', 'Cause Subcategory', 'Outcome of Incident']
for col in string_cols:
    df[col] = df[col].str.strip()

# 🧹 Step 5: Drop duplicate rows (if any)
df.drop_duplicates(inplace=True)

# ✅ Step 6: Check again after cleaning
print("\nAfter Cleaning:")
print(df.info())
print("\nSample Data:")
print(df.head())
df.to_csv("cleaned_road_accidents.csv", index=False)
```

## **Analyzing Accident Causes vs Outcomes** 


```python
#5. Analyzing Accident Causes vs Outcomes
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (already cleaned)
df = pd.read_csv("cleaned_road_accidents.csv")

# Clean just in case
df.dropna(subset=['Count'], inplace=True)
df['Count'] = df['Count'].astype(int)

# ✅ Create pivot table: rows = Cause category, columns = Outcome, values = Count
pivot_table = df.pivot_table(
    index='Cause category',
    columns='Outcome of Incident',
    values='Count',
    aggfunc='sum',
    fill_value=0
)

# ✅ Plot: Stacked bar chart
pivot_table.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab20')
plt.title("🛑 Accident Causes vs Outcomes")
plt.xlabel("Cause Category")
plt.ylabel("Total Number of Cases")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Outcome of Incident", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## **Accident Causes vs Outcomes (Stacked Bar):**

**Key Observations:**
- All cause categories show similar outcome distributions, dominated by total accidents and injuries.
- Fatalities and grievous injuries are consistently lower across all categories, but still present.

**Notable Patterns:**
- Weather and Traffic Violation categories contribute heavily to injury-related outcomes.
- 'Impacting Vehicle/Object' shows slightly lower total cases compared to systemic causes like Junction or Traffic Control.

**Implications:**
- Systemic factors (infrastructure, regulation) drive consistent accident severity across categories.
- Behavioral and environmental causes (e.g., violations, weather) amplify injury risks.

**Use Case:**
- Enables comparative severity analysis across accident causes.
- Supports targeted interventions by highlighting which causes lead to more harmful outcomes.
