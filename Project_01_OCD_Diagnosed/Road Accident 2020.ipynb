```python
#  Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#  Load the dataset
df = pd.read_csv(r"C:\Users\Admin\Documents\DataScience\Projects\Datasets\Regulatory Affairs of Road Accident Data 2020 India (1).csv")  

#  Data Cleaning
df.dropna(subset=['Count'], inplace=True)
df['Count'] = df['Count'].astype(int)

#  Check basic info (optional)
print("Unique Cities:", df['Million Plus Cities'].nunique())
print("Cause Categories:", df['Cause category'].unique())
print("Outcome Types:", df['Outcome of Incident'].unique())

#  1. Total Accidents by Cause Category
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Cause category', y='Count', estimator=sum, ci=None, palette='coolwarm')
plt.title("Total Accidents by Cause Category")
plt.xlabel("Cause Category")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#  2. Top 10 Cities by Total Accidents
top_cities = df.groupby('Million Plus Cities')['Count'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities.index, y=top_cities.values, palette='magma')
plt.title("Top 10 Million-Plus Cities by Total Accidents")
plt.xlabel("City")
plt.ylabel("Total Accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#  3. Heatmap: City vs Cause Category
pivot_table = df.pivot_table(index='Million Plus Cities', columns='Cause category', values='Count', aggfunc='sum', fill_value=0)

plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=0.5)
plt.title("Heatmap: City vs Cause Category")
plt.xlabel("Cause Category")
plt.ylabel("Million Plus Cities")
plt.tight_layout()
plt.show()

#  4. Outcome Distribution (All Cities Combined)
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Outcome of Incident', y='Count', estimator=sum, ci=None, palette='pastel')
plt.title("Distribution of Accident Outcomes in Million-Plus Cities (2020)")
plt.xlabel("Outcome")
plt.ylabel("Number of Cases")
plt.tight_layout()
plt.show()
```

```python

#Cleaning data
#  Import libraries
import pandas as pd

#  Load the dataset
df = pd.read_csv("/content/Regulatory Affairs of Road Accident Data 2020 India.csv")  # Replace with your actual filename

#  Step 1: View basic info
print("Before Cleaning:")
print(df.info())
print("\nMissing values:\n", df.isnull().sum())

#  Step 2: Drop rows where 'Count' is missing
df.dropna(subset=['Count'], inplace=True)

#  Step 3: Fix data types
df['Count'] = df['Count'].astype(int)

#  Step 4: Remove leading/trailing spaces in string columns
df.columns = df.columns.str.strip()
string_cols = ['Million Plus Cities', 'Cause category', 'Cause Subcategory', 'Outcome of Incident']
for col in string_cols:
    df[col] = df[col].str.strip()

#  Step 5: Drop duplicate rows (if any)
df.drop_duplicates(inplace=True)

#  Step 6: Check again after cleaning
print("\nAfter Cleaning:")
print(df.info())
print("\nSample Data:")
print(df.head())
df.to_csv("cleaned_road_accidents.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style="whitegrid")

# -------------------------------
# Load Cleaned Dataset
df = pd.read_csv("cleaned_road_accidents.csv")  # Replace if needed

# -------------------------------
# 1. Top 10 Cities with Most Accidents
city_counts = df.groupby("Million Plus Cities")["Count"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=city_counts.index, y=city_counts.values, palette="rocket")
plt.title("Top 10 Cities with Most Accidents (2020)")
plt.xlabel("Cities")
plt.ylabel("Total Accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# -------------------------------
# 2. Total Accidents by Cause Category
cause_counts = df.groupby("Cause category")["Count"].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=cause_counts.index, y=cause_counts.values, palette="mako")
plt.title("Accidents by Cause Category")
plt.xlabel("Cause Category")
plt.ylabel("Total Accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# -------------------------------
# 3. Most Common Subcategories (Top 10)
subcat_counts = df.groupby("Cause Subcategory")["Count"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=subcat_counts.index, y=subcat_counts.values, palette="coolwarm")
plt.title("Top 10 Accident Subcategories")
plt.xlabel("Cause Subcategory")
plt.ylabel("Total Accidents")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# -------------------------------
# 4. Distribution by Outcome Type
outcome_counts = df.groupby("Outcome of Incident")["Count"].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=outcome_counts.index, y=outcome_counts.values, palette="pastel")
plt.title("Distribution of Accident Outcomes")
plt.xlabel("Outcome Type")
plt.ylabel("Total Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# -------------------------------
# 5. Heatmap: Cause Category vs Outcome
pivot = df.pivot_table(index='Cause category', columns='Outcome of Incident', values='Count', aggfunc='sum', fill_value=0)

plt.figure(figsize=(14, 8))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=0.5)
plt.title("Heatmap: Cause Category vs Outcome")
plt.xlabel("Outcome of Incident")
plt.ylabel("Cause Category")
plt.tight_layout()
plt.show()


```

```python
#B. Analysis of Accident Causes
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv("cleaned_road_accidents.csv")

#  Step 1: Clean if not already cleaned
df.dropna(subset=['Count'], inplace=True)
df['Count'] = df['Count'].astype(int)

#  Step 2: Aggregate accident counts by cause category
cause_category = df.groupby('Cause category')['Count'].sum().sort_values(ascending=False)

# Plot: Cause Category
plt.figure(figsize=(12, 6))
sns.barplot(x=cause_category.values, y=cause_category.index, palette='viridis')
plt.title(" Total Accidents by Cause Category")
plt.xlabel("Total Number of Accidents")
plt.ylabel("Cause Category")
plt.tight_layout()
plt.show()

#  Step 3: Aggregate accident counts by cause subcategory
top_subcat = df.groupby('Cause Subcategory')['Count'].sum().sort_values(ascending=False).head(15)

# Plot: Top 15 Cause Subcategories
plt.figure(figsize=(12, 6))
sns.barplot(x=top_subcat.values, y=top_subcat.index, palette='magma')
plt.title(" Top 15 Accident Cause Subcategories")
plt.xlabel("Total Number of Accidents")
plt.ylabel("Cause Subcategory")
plt.tight_layout()
plt.show()
```

```python
#C. Outcomes of Incidents
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the already-cleaned dataset (replace with your correct path if needed)
df = pd.read_csv("cleaned_road_accidents.csv")

# Clean just in case
df.dropna(subset=['Count'], inplace=True)
df['Count'] = df['Count'].astype(int)

#  Group data by 'Outcome of Incident'
outcome_counts = df.groupby('Outcome of Incident')['Count'].sum().sort_values(ascending=False)

#  Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=outcome_counts.values, y=outcome_counts.index, palette="Set2")
plt.title(" Distribution of Road Accident Outcomes", fontsize=14)
plt.xlabel("Total Number of Cases")
plt.ylabel("Outcome of Incident")
plt.tight_layout()
plt.show()
```

```python
#5. Analyzing Accident Causes vs Outcomes
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (already cleaned)
df = pd.read_csv("cleaned_road_accidents.csv")

# Clean just in case
df.dropna(subset=['Count'], inplace=True)
df['Count'] = df['Count'].astype(int)

#  Create pivot table: rows = Cause category, columns = Outcome, values = Count
pivot_table = df.pivot_table(
    index='Cause category',
    columns='Outcome of Incident',
    values='Count',
    aggfunc='sum',
    fill_value=0
)

#  Plot: Stacked bar chart
pivot_table.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab20')
plt.title(" Accident Causes vs Outcomes")
plt.xlabel("Cause Category")
plt.ylabel("Total Number of Cases")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Outcome of Incident", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
